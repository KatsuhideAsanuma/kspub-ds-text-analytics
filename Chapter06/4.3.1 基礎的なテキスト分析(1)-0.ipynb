```python
# read_review1.py

import pandas as pd

# 宮島の口コミのファイルの読み込み
df = pd.read_csv('text/miyajima_sample.csv', encoding='utf-8',
                 parse_dates=['旅行の時期'])
display(df)
```

```python
# read_review2.py

# 口コミの件数の算出　
count = len(df)
print("口コミの件数", count)

# 口コミの平均文字数の算出
word_count = 0
for review in df["口コミ"]:
    word_count += len(review)

ave = word_count / count
print("口コミの平均文字数", round(ave, 1))
```
